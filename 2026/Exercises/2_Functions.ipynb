### Recursive functions [Difficult]

A recursive function is a function that calls itself until it reaches a stopping condition, called the base case.

Here is an example of the structure : 

```python
def recursive_function(n):
    if n == base_case:      # Stop condition
        return some_value
    return recursive_function(n - 1)  # Self-call, moving toward base case
```

Key Rules <br>
- Always define a base case, otherwise the function loops forever
- Each call must get closer to the base case
- Python has a default recursion limit of 1000 calls
