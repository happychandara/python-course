### Exercise 7 - Check if a list is sorted

Create a function that determines if a list of numbers is sorted in non-decreasing (ascending) order. Return True if it is, and False otherwise. Test it with [1, 2, 6, 8, 9] and [2, 7, 4, 6]. 

```python
def check_sorted(list):
    if len(list) <= 1 : 
        return True
    else : 
        for i in range(1, len(list)):
            if list[i-1] > list[i] : 
                return False
        return True
    
print(check_sorted([1, 2, 6, 8, 9]))
print(check_sorted([2, 7, 4, 6]))
```

Remove all duplicate values from the list ['a', 'b', 'a', 'c', 'b'] while preserving the original order of the first occurrence.

```python
list = ["a", "b", "a", "c", "b"]
no_duplicates = []
for letter in list : 
    if letter not in no_duplicates:
        no_duplicates.append(letter)
print(no_duplicates)
```

### Exercise 9 - Rotate a list

Create a function that rotates a list by a given number of steps to the right. For a list [1, 2, 3, 4, 5] and k=2, the result should be [4, 5, 1, 2, 3]. Test it on the list [1, 2, 3, 4, 5, 6, 7, 8] for k=3 and k=11.

```python
def rotate(list, k):
    if k == 0 : 
        return list
    if k > len(list):
        k = k % len(list)
    return list[-k:] + list[:-k]

print(rotate([1, 2, 3, 4, 5, 6, 7, 8], 11))
```

### Exercise 3 - Dictionary from 2 lists

Write a function to create a dictionary by mapping two equal-length lists, one containing keys and the other containing values. Try it with  keys = ["name", "age", "city"] and values = ["Bob", 25, "London"]. it should give you {'name': 'Bob', 'age': 25, 'city': 'London'}. 

```python
def map_dico(keys, values):
    if len(keys) != len(values): 
        return None
    else : 
        answer = {}
        for i in range(len(keys)):
            answer[keys[i]] = values[i]
        return answer
    
print(map_dico(["name", "age", "city"], ["Bob", 25, "London"]))
```

### Exercise 4 - Merge 2 dictionaries

Write a function to combine two dictionaries into a single dictionary. If both dictionaries share a key, the value from the second dictionary should take precedence. Test it with dict1 = {"a": 1, "b": 2} and dict2 = {"b": 3, "c": 4}.

Purpose: Merging dictionaries is a frequent task in configuration management, data aggregation, and API response handling, making it essential to know the available approaches and their precedence rules.

```python
def merge_dico(dico1, dico2):
    answer = dico2
    for key in dico1:
        if key not in answer:
            answer[key] = dico1[key]
    return answer

print(merge_dico({"a":1, "b":2}, {"b":3, "c":4}))
```

### Exercise 6 - Check key and value existence

Write two functions : 
- one that checks if a key is in the dictionary
- one that checks if a value is in the dictionary


Test your functions with the dictionary roles = {"alice": "admin", "bob": "editor", "carol": "viewer"} : 
- test with the keys "bob" and "viewer". It should return True and False
- test with the values "bob" and "viewer". It should return False and True

```python
def key_in_dico(dico, key):
    return key in dico

def value_in_dico(dico, value):
    for key in dico : 
        if dico[key] == value :
            return True
    return False

roles = {"alice":"admin", "bob":"editor", "carol":"viewer"}
print(key_in_dico(roles, "bob"))
print(key_in_dico(roles, "viewer"))
print(value_in_dico(roles, "bob"))
print(value_in_dico(roles, "viewer"))
```

### Exercise 7 - Letter frequency 

Write a function that returns a dictionary of the letter frequencies of a string. For instance, for the input `hello`, the function will return `{"h": 1, "e": 1, "l": 2, "o": 1}`. Test it with `hello world`. 

```python
def frequency(string):
    ans = {}
    for letter in string : 
        if letter in ans : 
            ans[letter] += 1
        else : 
            ans[letter] = 1
    return ans

print(frequency("hello world"))
```

### Exercise 8 - Convert values to int

Write a function to convert string values of a given dictionary into integer datatypes. Test it with {'x': '10', 'y': '20', 'z': '30'}. The output should be {'x':10, 'y':20, 'z':30}.

```python
def to_int(dict):
    for key in dict:
        dict[key] = int(dict[key])
    return dict

print(to_int({'x': '10', 'y': '20', 'z': '30'}))
```

### Exercise 9 - Combinations of keys and values

From a dictionary, create a function that returns all the combinations possible of keys and values under a list of tuples. For instance : 
- Input : {"species": "cat", "age": 10, "name": "Semsem"}
- Output : [("species", "cat"),("species", 10), ("species", "Semsem"), ("age", "cat"), ("age", 10), ("age", "Semsem"), ("name", "cat"), ("name", 10), ("name", "Semsem")]

```python
def combinations(dico):
    ans = []
    for key in dico.keys():
        for value in dico.values():
            ans.append((key, value))
    return ans

print(combinations({"species": "cat", "age": 10, "name": "Semsem"}))
```
