### Exercise 5 - Convert a list of strings to uppercase

We use the function `word.upper()` to put the word "word" to uppercase : "WORD". Create a function `to_uppercase` which takes a list of words as argument, and returns this same list but with each word in uppercase. Try it with the list ["cat", "dog", "bird"]. 

```python
def to_uppercase(words):
    result = []
    for word in words:
        result.append(word.upper())
    return result

words = ["cat", "dog", "bird"]
print(to_uppercase(words))
```

### Exercise 6 - Calculate the average of a list

Create a function that calculates the average of a list of numbers. It should take a list as input, and return the average as output. 

```python
def calculate_average(numbers):
    if len(numbers) == 0:
        return 0  # avoid division by zero
    total = 0
    for n in numbers:
        total += n
    return total / len(numbers)

# Example
values = [10, 15, 20, 25]
print(calculate_average(values))
```

### Exercise 7 - Sum two matrices [Difficult]

Create 2 matrices : 
- `M` , a matrix 3X3 that takes the numbers 1 to 9 included in ascending order
- `N`, a matrix 3X3 that takes the numbers 9 to 1 included in descending order

Create the matrix `S`, which is the addition of the matrices `M` and `N` : the coefficient with coordinates $(1,1)$ of the matrix `M` is added to the coefficient with the same coordinates of the matrix `N`, and gives the coefficient $(1,1)$ of the matrix `S`.

```python
M = [[1, 2, 3],[4, 5, 6],[7, 8, 9]]
N = [[9, 8, 7],[6, 5, 4],[3, 2, 1]]
S = [[0]*3 for _ in range(3)]

for i in range(len(M)):
    for j in range(len(M[0])):
        S[i][j] = M[i][j] + N[i][j]

print(S)
```

### Exercise 8 - Substrings

Write a function `get_string` that returns the first 3 characters of a string.

```python
def get_string(word):
    if len(word) < 3:
        return word
    else : 
        return word[:3]

print(get_string("no"))
print(get_string("Europe"))
```

### Exercise 11 - Count elements

Write a function `count_elements` that counts how many times each element appears in a list and returns a dictionary of the result, with the key being the element, and the value the number of times it appears. Test it with the list ["a", "b", "a", "c", "d", "a", "d", "a"].

```python
def count_elements(list):
    counts = {}
    for item in list:
        if item in counts:
            counts[item] += 1
        else:
            counts[item] = 1
    return counts

count_elements(["a", "b", "a", "c", "d", "a", "d", "a"])
```

### Exercise 12 - Concatenate dictionaries

Write a function `merge_dicts` that merges two dictionaries. If a key exists in both, add their values. Test it with {"a": 1, "b": 2} and {"b": 3, "c": 4}.

```python
def merge_dicts(d1, d2):
    result = d1.copy()

    for key in d2:
        if key in result:
            result[key] += d2[key]
        else:
            result[key] = d2[key]

    return result

print(merge_dicts({"a":1, "b":2}, {"b":3, "c":4}))
```
