### Exercise 5 - Product of matrixes [Difficult]

Given two Numpy arrays (matrices), write a function to compute the product of the two matrices of shape 2X2 (not the product coefficient by coefficient). As a reminder, for 2 matrices [[a, b], [c, d]] and [[e, f], [g, h]], the product of the matrices [[ae+bg, af+bh], [ce+dg, cf+dh]]. 
Test your function. 

```python
def product(M, N):
    S = np.zeros((2,2))
    for i in range(2):
        for j in range(2):
            S[i, j] = M[i, 0]*N[0, j] + M[i, 1]*N[1, j]
    return S


print(product(np.array([[1, 2],[3, 4]]), np.array([[0, 1],[2, 3]])))
```

### Exercise 7 - Quadratic function explorer

Given the quadratic function:
$f(x)=2x^2−3x−5$
Tasks:
1. Create an array x of 500 values between -3 and 5 using np.linspace
2. Compute f(x) for all values at once
3. Find the minimum value of f(x) and the x where it occurs
4. Verify the two roots are close to x = -1 and x = 2.5. find the x values in your array closest to each root

Hint : you can use `np.argmin(x)` to find the index of the minimum of the list. 

```python
# 1. x values
x = np.linspace(-3, 5, 500)

# 2. Compute f(x)
f = 2*x**2 - 3*x - 5

# 3. Minimum of f(x)
min_val = np.min(f)
min_x   = x[np.argmin(f)]
print(f"Minimum: f({min_x:.3f}) = {min_val:.3f}")
# Minimum: f(0.751) = -6.125

# 4. Where is f(x) < 0?
below = x[f < 0]
print(f"f(x) < 0 from x = {below[0]:.3f} to x = {below[-1]:.3f}")
# f(x) < 0 from x = -1.000 to x = 2.500

# 5. Closest x to each root
for root in [-1.0, 2.5]:
    idx = np.argmin(np.abs(x - root))
    print(f"Near root {root}: f({x[idx]:.4f}) = {f[idx]:.4f}")
# Near root -1.0:  f(-1.0000) = 0.0000
# Near root  2.5:  f( 2.5000) = 0.0000
```
