### Exercise 4 - Fibonacci sequence

The Fibonacci sequence is a sequence of numbers, defined as follows : 
$F_{n+2}=F_{n+1} + F_n$, with $F_0 = 1$ and $F_1 = 1$.
Create an algorithm that displays the `n` first numbers of this sequence and test it with `n=5` and `n=10`.

```python
n = 5
a = 1
b = 1
print(a)
print(b)
for i in range(n):
    a, b = b, a+b
    print(b)
```

### Exercise 6 - Guess the number

Write a program that makes the player guess a number between 1 and 10. Every time the player is wrong, display "Wrong number. Try again!". When the player is right, display "You guessed the number, well done!" and end the program. 

Hint: you can use the function `random.randint(a,b)` that chooses a random number between `a` and `b`. 
You can also use the function `input(Sentence of the input)` to collect the input number from the player. 

```python
import random

guess_nb = random.randint(1, 11)
input_nb = int(input("Try a number : "))
while input_nb != guess_nb:
    print("Wrong number. Try again!")
    input_nb = int(input("Try a number : "))
print("You guessed the number, well done!")
```

### Exercise 8 - Use a function to improve readability

Here is an algorithm very repetitive, that could be optimized by creating a function. Try to re-write the algorithm using a function. Ensure your function works. 

```python
number = 32
n, m = number, number
while n % 3 == 0 : 
    n = n // 3 
if n == 1 : 
    print(number, "is a power of 3")
else : 
    print(number, "is not a power of 3")
while m % 2 == 0 : 
    m = m // 2
if m == 1 : 
    print(number, "is a power of 2")
else : 
    print(number, "is not a power of 2")
```

```python
def isPowerOf(input, multiple):
    number = input
    while number % multiple == 0 : 
        number = number // multiple
    if number == 1 : 
        print(input, "is a power of", multiple)
    else : 
        print(input, "is not a power of", multiple)

isPowerOf(32, 3)
isPowerOf(32, 2)
```
