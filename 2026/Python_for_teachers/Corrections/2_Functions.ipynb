Below is a very repetitive algorithm. Try to make it less repetitive by using a function for the repeated part. 

```python
# A very repetitive algorithm. 
square1 = 1**2
cube1 = 1**3
print("x =", 1, "square =", square1, "cube =", cube1)

square2 = 2**2
cube2 = 2**3
print("x =", 2, "square =", square2, "cube =", cube2)

square3 = 3**2
cube3 = 3**3
print("x =", 3, "square =", square3, "cube =", cube3)

square4 = 4**2
cube4 = 4**3
print("x =", 4, "square =", square4, "cube =", cube4)
```

### Exercise 6 - Function using for loop

Write a Python function that computes the sum of squares of the first n integers.
Mathematically, this means:
$S(n)=1^2+2^2+3^2+⋯+n^2$. Try it with $n=10$.

```python
def sum_squares(n):
    sum = 0
    for i in range(1,n+1):
        sum = sum + i**2
    return sum

print(sum_squares(10))
```

### Exercise 9 - Quadratic function

Build a `quadratic_solver` function that solves a quadratic equation of the form ax² + bx + c = 0 using the quadratic formula, where:
- a, b, c (required) — the coefficients
- round_digits=2 — decimal precision (optional)

Reminder : the discriminant is b² − 4ac. If the discrimant is : 
- positive : print 2 solutions,  x = (−b ± √(b² − 4ac)) / 2a
- null : print 1 solution, x = x = −b / 2a
- negative : return None

Test your function with `a=2, b=1, c=1`, then `a=-20, b=1, c=1, round_digits=2`, then `a=-20, b=1, c=1, round_digits=1` and `a=1, b=2, c=1`

```python
def quadratic_solver(a, b, c, round_digits=2):
    discriminant = b**2 - 4*a*c

    if discriminant < 0:
        return None
    elif discriminant == 0: 
        x = -b / (2*a)
        return round(x, round_digits)
    else : 
        x1 = (-b + discriminant**0.5) / (2*a)
        x2 = (-b - discriminant**0.5) / (2*a)
        return (round(x1, round_digits), round(x2, round_digits))
    
print(quadratic_solver(2,1,1))
print(quadratic_solver(-20,1,1, 2))
print(quadratic_solver(-20,1,1, 1))
print(quadratic_solver(1, 2, 1))
```

### Recursive functions [Difficult]

A recursive function is a function that calls itself until it reaches a stopping condition, called the base case.

Here is an example of the structure : 

```python
def recursive_function(n):
    if n == base_case:      # Stop condition
        return some_value
    return recursive_function(n - 1)  # Self-call, moving toward base case
```

Key Rules <br>
- Always define a base case, otherwise the function loops forever
- Each call must get closer to the base case
- Python has a default recursion limit of 1000 calls

### Exercise 10 - Recursive factorial

Build a factorial function that computes $n!$ recursively.
Formula: $n! = n × (n−1) × ... × 1$, and $0! = 1$
Try it with $n=8$.

```python
def factorial(n):
    if n == 0:         
        return 1
    return n * factorial(n - 1)

print(factorial(8))
```

### Exercise 11 - Fibonacci factorial

Build a Fibonacci function that returns the nth Fibonacci number recursively. Remind that the Fibonacci sequence is defined as following : 
$F_{n+2}=F_{n+1}+F_n, F_1 = 1, F_0 = 0$.

```python
def fibonacci(n):
    if n == 0 : 
        return 0
    elif n == 1 : 
        return 1
    else : 
        return fibonacci(n-1)+fibonacci(n-2)
    
print(fibonacci(10))
```
