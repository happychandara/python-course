# Tic tac toe game

The purpose of this lesson is to create a tic tac toe game, using lists. 

## Correction

### Stage 1 - Build and display the grid

The board is a list of 9 cells. Each cell has the state "-" at the beginning of the game, and turns into "O" or "X", depending of the players' choice. 

1. Create the function `create_board`, that creates the board as a list of 9 elements, all equal to "-". The function should return the board. 
2. Complete the function `display_board`, that displays the board. 
3. Call the 2 functions to test them. 

```python
def create_board():
    board = ["-", "-", "-",
             "-", "-", "-",
             "-", "-", "-"]
    return board

def display_board(board):
    print()
    print(f" {board[0]} | {board[1]} | {board[2]}")
    print("---+---+---")
    print(f" {board[3]} | {board[4]} | {board[5]}")
    print("---+---+---")
    print(f" {board[6]} | {board[7]} | {board[8]}")
    print()

# Test it
board = create_board()
display_board(board)
```

### Stage 2 - Place a mark

Complete the function `place_mark`, with the necessary conditions to check that the position is between 1 and 9, and if the cell is already taken before modifying it. The parameter `board` is the list created before, and the `player` is either "X" or "O". 

Then, test it by creating the board and trying to place a mark.

```python
def place_mark(board, player):
    while True : 
        position = int(input(f"Player {player}, choose a position (1-9): "))

        if position < 1 or position > 9:
            print("Position must be between 1 and 9.")
            continue

        index = position - 1
        if board[index] != "-":
            print("This cell is already taken. Choose another.")
            continue

        board[index] = player
        return


# Test it
board = create_board()
display_board(board)
place_mark(board, "X")
display_board(board)
```

### Stage 3 - Check for a winner

1. Now, we will check if someone won, by checking if some combinations of 3 positions have the same elements (diagonals, rows, columns). For this, complete the winning_combinations in the function `check_winner`. 
2. Create the function `check_draw`, that checks if the game is finished. The function takes the board as a parameter, and returns True if the game is finished, and False otherwise. 
3. Test both functions with appropriate boards. 

```python
def check_winner(board):
    winning_combinations = [
        [0, 1, 2],  
        [3, 4, 5],  
        [6, 7, 8], 
        [0, 3, 6],  
        [1, 4, 7],  
        [2, 5, 8], 
        [0, 4, 8],  
        [2, 4, 6], 
    ]

    for combo in winning_combinations:
        a = combo[0]
        b = combo[1]
        c = combo[2]
        if board[a] == board[b] == board[c] and board[a] != "-":
            return board[a]   # return "X" or "O"

    return None   # no winner yet

def check_draw(board):
    for cell in board:
        if cell == "-":
            return False   # at least one cell is still empty
    return True   # all cells filled, no winner

# Test it
board = ["X", "X", "X",
         "-", "O", "-",
         "-", "-", "O"]
winner = check_winner(board)
print(winner)  
```

### Stage 4 - Switch players

The player should switch between "O" and "X" each time a new item is drawn on the board. Create the function `switch_player`, that takes current_player as a parameter, and returns the new player (either "O" or "X"). 

```python
def switch_player(current_player):
    if current_player == "X":
        return "O"
    else:
        return "X"
```

### Stage 5 - The game loop

Finally, complete the function `play_game`, that uses alll the previous functions to make the game complete. In this function, the players alternate turns until someone wins or the board is full.

```python
def play_game():
    board = create_board()
    current_player = "X"

    print("=== TIC-TAC-TOE ===")

    game_over = False
    while not game_over:
        display_board(board)
        place_mark(board, current_player)

        winner = check_winner(board)
        if winner:
            display_board(board)
            print(f"Player {winner} wins!")
            game_over = True

        elif check_draw(board):
            display_board(board)
            print("It's a draw!")
            game_over = True

        else:
            current_player = switch_player(current_player)

play_game()
```
