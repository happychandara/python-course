### Exercise 1 - Velocity vector of a material point


The script below is used to display the trajectory of a moving point and the average velocity vector of that point on a given date.

1. Recall the formula for the velocity vector at point $M_3$, approximated by the average velocity vector between $M_2$, with coordinates $(x_2, y_2)$ and $M_3$, with coordinates $(x_3, y_3)$. Which difference can be observed in the calculation of the velocity vector’s coordinates in the script?

$\vec{v_{3}} = (x_3 - x_2,y_3 - y_2)$

```python
def velocity(x,y,i) :
  # x and y are the lists of the coordinates and i is the index of the point we want to calculate. 
  plt.arrow(x[i],y[i],(x[i+1]-x[i]),(y[i+1]-y[i]) , shape='full', lw=1,
  length_includes_head=True, rasterized=True, color='r', head_width=.05,fc='r')
```

2. The following code displays the different positions of the point. Try to use the velocity function to display the velocity on each point as well. 

```python
# Define the positions of the point. 
x = np.linspace(0, 2.4, 9)
y = np.array([0, 0.25, 0.4, 0.46, 0.42, 0.28, 0.04, -0.3, -0.74])

# Display the positions
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

# Display the velocity
for i in range(len(x)-1):
    velocity(x, y, i)

plt.show()
```

3. Create the function `velocity_variation`, which displays on the graph the variation $\Delta \vec{v_i} = \vec{v_{i+1}} - \vec{v_i}$ between 2 consecutive velocities. It should take x, y and i as an argument. 

```python
def velocity_variation(x, y, i):
    plt.arrow(x[i],y[i],(x[i+2]-2*x[i+1]+x[i]),(y[i+2]-2*y[i+1]+y[i]) , shape='full', lw=1,
  length_includes_head=True, rasterized=True, color='g', head_width=.03,fc='g')
```

```python
# Define the positions of the point. 
x = np.linspace(0, 2.4, 9)
y = np.array([0, 0.25, 0.4, 0.46, 0.42, 0.28, 0.04, -0.3, -0.74])

# Display the positions
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

# Display the velocity
for i in range(len(x)-1):
    velocity(x, y, i)

# Display the velocity variation
for i in range(len(x)-2):
    velocity_variation(x, y, i)

plt.show()
```

### Exercise 2 : Simulating projectile motion

When an object is thrown upward or at an angle, its vertical position changes according to:
$$ y = y_0 + v_0t - \frac{1}{2}gt^2$$

where : 
- $y_0$ = initial height 
- $v_0$ = initial vertical velocity
- $g = 9.81 m/s^2$
- $t$ = time

1. Generate a list of 41 elements, that represents times from 0 to 4 seconds. Why is it useful to have many time values ? 

```python
t = np.linspace(0, 4, 41)
```

2. Suppose a ball is launched vertically from the ground with:
- Initial height: $y_0 = 0$ m
- Initial velocity: $v_0 = 20$ m/s

Compute its height over time and print it. 

```python
# Initializes
g = 9.81
y0 = 0
v0 = 20

# Compute its height
y = y0 + v0 * t - 0.5 * g * t**2

print(y)
```

3. 
- What is the height at t=0?
- Does the height always increase?
- At approximately what time does the ball start descending?

- The height is 0 at $t=0$
- No, it starts going down because of the gravity after some time. 
- At the index 21, which corresponds 2.1 s. 

4. Use NumPy to determine the highest point reached.

```python
max_height = np.max(y)
index = np.argmax(y)
time_at_max = t[index]

print("Maximum height:", max_height)
print("Time at maximum height:", time_at_max)
```

Visualize the trajectory of the ball for $v_0=15$ m/s and $v_0=25$ m/s, by running the next cell. Try to answer those questions : 
- Which launch reaches the greater height?
- How does doubling the launch speed affect the trajectory?
- Which ball remains in the air longer?

```python
plt.plot(t, 15 * t - 0.5 * g * t**2, label="v0 = 15 m/s")
plt.plot(t, 25 * t - 0.5 * g * t**2, label="v0 = 25 m/s")

plt.xlabel("Time (s)")
plt.ylabel("Height (m)")
plt.title("Projectile Motion")
plt.legend()
plt.grid(True)

plt.show()
```
