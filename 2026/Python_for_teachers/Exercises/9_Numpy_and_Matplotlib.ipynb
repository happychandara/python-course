### Exercise 1 - Exploring sine waves

A sine wave is described by the equation : 
$$y=Asin(nx)$$

where:
- A is the amplitude
- n determines the frequency (or harmonic number).

The <b>amplitude</b> is the maximum distance between the center line and the peak of the wave. Changing the amplitude makes the wave taller or shorter.

The <b>period</b> is 1/frequency. The <b>period</b> is the horizontal distance required for one complete cycle, whereas the <b>frequency</b> tells us how many complete cycles occur over a given interval.
