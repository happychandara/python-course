1. Complete the following code to display "You are tall" if the height is superior to 170cm. Display "You are small" if the height is inferior to 140cm. Otherwise, display "Your size is normal". Try the program with different sizes.

### Exercice 3 - While loop

1. Complete the program to display all the numbers from 1 to 15 included, this time using a while loop instead of a for loop. Display "Done" at the end. 

```python
n = 1
while n <= 15 : 
    print(n)
    n = n + 1
print("Done")
```

2. A company take a loan of 3000 $ at the bank on year 0. Suppose that they don't pay their loan back, and that each year, the 3000 dollars increase by 5%, due to interest rates. Calculate the amount they have to pay to the bank in order to pay back their loan after 15 years, only using a while loop. 

```python
price = 3000
year = 0 
while year <= 15 : 
    price = price*1.05
    year += 1
print("The price after 15 years is", price)
```

### Exercise 6 - Understand the use of a function

Below is a very repetitive algorithm. Try to make it less repetitive by using a function for the repeated part. 

```python
# A very repetitive algorithm. 
square1 = 1**2
cube1 = 1**3
print("x =", 1, "square =", square1, "cube =", cube1)

square2 = 2**2
cube2 = 2**3
print("x =", 2, "square =", square2, "cube =", cube2)

square3 = 3**2
cube3 = 3**3
print("x =", 3, "square =", square3, "cube =", cube3)

square4 = 4**2
cube4 = 4**3
print("x =", 4, "square =", square4, "cube =", cube4)
```

```python
# The function 
def operations(x):
    square = x**2
    cube = x**3
    print("x =", x, "square =", square, "cube =", cube)

# The code using the function
operations(1)
operations(2)
operations(3)
operations(4)
```
